### Import relevant packages

```python
import SimpleITK as sitk
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

### Read in the images
We want to read in the "head_fix.png" and "head_mov.png" images so that we can perform rigid translational registration.

```python
file1 = 'ITK_Images/head_fix.png'
file2 = 'ITK_Images/head_mov.png'
cost = []
pos = []; posy = []; posx = [];

fixed = sitk.ReadImage(file1, sitk.sitkFloat32)
moving = sitk.ReadImage(file2, sitk.sitkFloat32)
# command that will allow us to print and save our cost function as it evolves.
def command_iteration(method):
    print(f"Iteration{method.GetOptimizerIteration():3} : Cost = {method.GetMetricValue():10.5f} : {method.GetOptimizerPosition()}")
    cost.append(method.GetMetricValue())
    [x, y] = method.GetOptimizerPosition()
    posx.append(x)
    posy.append(y)
```

### Set the Registration method and Execute
Now that we have loaded in the images, lets setup and perform the registration:

Our metric to evaulate our registration: Mean squares  
Our algorithm for optimization: Gradient Descent with a learning rate of 4.0, 200 iterations, and min step of 0.01   
Our transform: Rigid translation  
Interpolator: Linear interpolation

```python
R = sitk.ImageRegistrationMethod()
R.SetMetricAsMeanSquares()
R.SetOptimizerAsRegularStepGradientDescent(4.0, .01, 200)
R.SetInitialTransform(sitk.TranslationTransform(fixed.GetDimension()))
R.SetInterpolator(sitk.sitkLinear)
R.AddCommand(sitk.sitkIterationEvent, lambda: command_iteration(R))

outTx = R.Execute(fixed, moving)
print("-------")
print(outTx)
print(f"Optimizer stop condition: {R.GetOptimizerStopConditionDescription()}")
print(f" Iteration: {R.GetOptimizerIteration()}")
print(f" Metric value: {R.GetMetricValue()}")
cost.append(R.GetMetricValue())
resampler = sitk.ResampleImageFilter()
resampler.SetReferenceImage(fixed)
resampler.SetInterpolator(sitk.sitkLinear)
resampler.SetDefaultPixelValue(100)
resampler.SetTransform(outTx)

out = resampler.Execute(moving)
simg1 = sitk.Cast(sitk.RescaleIntensity(fixed), sitk.sitkUInt8)
simg2 = sitk.Cast(sitk.RescaleIntensity(out), sitk.sitkUInt8)
cimg = sitk.Compose(simg1, simg2, simg1 // 2. + simg2 // 2.)
```

### Now lets plot the results of the registration
Above we have shown printed the translated value along with its associated cost function. Now we can show the results compared to the originals.

```python
fig, axs = plt.subplots(1,3,figsize = (15,15))
img_fix = sitk.GetArrayFromImage(fixed)
img_mov = sitk.GetArrayFromImage(moving)
img_out = sitk.GetArrayFromImage(out)

axs[0].imshow(img_fix,cmap='gray'); axs[0].axis('off'); axs[0].set_title('Fixed',fontsize=16)
axs[1].imshow(img_mov,cmap='gray'); axs[1].axis('off'); axs[1].set_title('Moving',fontsize=16)
axs[2].imshow(img_out,cmap='gray'); axs[2].axis('off'); axs[2].set_title('Translated',fontsize=16)
plt.show()
```

### Show difference image and cost function evolution
The registration above looks fairly good, but how does it compared to the fixed image?  
Lets show the difference image initially, the final difference (after translation), the cost function evolution, and finally, the translation position evolution.


```python
fig, axs = plt.subplots(2,2,figsize = (10,10))
diff_init = img_fix-img_mov
diff_final = img_fix-img_out
its = np.linspace(1,R.GetOptimizerIteration(),num=R.GetOptimizerIteration())


axs[0,0].imshow(diff_init,cmap='gray'); axs[0,0].axis('off'); axs[0,0].set_title('Initial Difference',fontsize=14)
axs[0,1].imshow(diff_final,cmap='gray'); axs[0,1].axis('off'); axs[0,1].set_title("Final Difference",fontsize=14)
axs[1,0].plot(its,cost,linewidth=3.0); axs[1,0].set_xlabel('Iteration',fontsize=14); axs[1,0].set_ylabel('Cost function',fontsize=14)
axs[1,0].set_title('Cost function evolution',fontsize=14)
axs[1,1].plot(its[:-1], posx,label='X shift')
axs[1,1].plot(its[:-1], posy,label='Y shift');
axs[1,1].set_title('Translation evolution',fontsize=14);axs[1,1].set_xlabel('Iteration',fontsize=14);
axs[1,1].set_ylabel("Translation",fontsize=14); axs[1,1].legend()
plt.show()
```
