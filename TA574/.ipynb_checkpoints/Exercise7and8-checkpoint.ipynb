## Import packages

```python
import os
import numpy as np
import skimage as sk
from matplotlib import pyplot as plt
from skimage import io, filters, exposure, segmentation
import SimpleITK as itk
```

## Set Path to image 
Set your path to the appropriate image in the cell below

```python
image_path = 'ITK_Images/'
img = 'BrainProtonDensitySlice6.png'
filename = os.path.join(image_path,img)
```

## Read in image and display

```python
brain = io.imread(filename)
I0 = plt.imshow(brain,cmap = 'gray');
brain_filt = np.full_like(brain, 0)
brain_filt = brain_filt.astype(float)
```

## Preprocessing
### 1. Reduce Noise
We will reduce noise by using an anisotropic diffusion filter

```python
CADIF = itk.CurvatureAnisotropicDiffusionImageFilter()
CADIF.SetTimeStep = 0.125
CADIF.SetNumberOfIterations = 5
CADIF.SetConductanceParameter = 9.0
CADIF.SetInput = brain.astype(float)
CADIF.Execute(brain_filt)
#brain_filt = brain_filt.astype(float)
#I1 = plt.imshow(brain_filt,cmap='gray');
```

```python
help(CADIF)
```

### 2. Create a gradient image
We will create a gradient image using a Sobel filter on the smoothed image

```python
brain_edge = filters.sobel(brain_filt)
brain_edge = brain_edge.astype(float)
I2 = plt.imshow(brain_edge,cmap='gray');
```

### 3. Normalize magnitude
Use a sigmoid curve to normalize the gradient magnitude image.

```python
def sigmoid_filter(input, s_max, s_min, alpha, beta):
    output = 1+np.exp(-1*(input-beta)/(alpha))
    output = (s_max-s_min)* output + s_min
    return output

#brain_norm = sigmoid_filter(brain_edge,1.0, 0.0, -0.1, 2.0)
brain_norm = exposure.adjust_sigmoid(brain_edge, cutoff = 0.3,gain = 0.1, inv=True)
I3 = plt.imshow(brain_norm,cmap='gray');
```

```python
s = np.linspace(0, 2*np.pi, 100)
r = 110 + 10*np.sin(s)
c = 50 + 10*np.cos(s)
init = np.array([r, c]).T
snake = segmentation.morphological_geodesic_active_contour(brain_norm,iterations = 500)

# Plot
fig, ax = plt.subplots(figsize=(7, 7))
ax.imshow(brain_norm, cmap=plt.cm.gray)
ax.plot(snake[:, 1], snake[:, 0], '-b', lw=3)
ax.set_xticks([]), ax.set_yticks([])
ax.axis([0, brain_norm.shape[1], brain_norm.shape[0], 0])
plt.show()
```
