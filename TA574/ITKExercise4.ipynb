### Import Packages
Here we will be using SimpleITK again. This library has multiple demons registation classes available to use.
First we will: 
- Import appropriate packages
- Load in two liver images at different points in the respiratory cycle
- Define a function that will allow us to track the metric value with each iteration.

```python
import SimpleITK as sitk
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

file1 = 'ITK_Images/liver_exp.png'
file2 = 'ITK_Images/liver_insp.png'
fixed = sitk.ReadImage(file1, sitk.sitkFloat32)
Moving = sitk.ReadImage(file2, sitk.sitkFloat32)
moving = Moving
def command_iteration(filter,print_flag):
    if print_flag == 1:
        print(f"{filter.GetElapsedIterations():3} = {filter.GetMetric():10.5f}")
    cost.append(filter.GetMetric())


print(moving.GetHeight())
print(moving.GetWidth())
print(moving.GetDepth())
```

### Demons Registration
Demons is a form of Difformable image registration (DIR). I have included SITK's "classic" Demons filter as well as the newer version that implements demons with "Fast Symmetric Forces". This is controlled with the "DEMON_flag".  
  
***Some History:*** It was first introduced to the medical community in 1995/1998 by JP Thirion, where the approach was to model image registration as a diffusion process. More specifically, its name comes from 19th century physicist, James Clark Maxwell's paradoxical thermodynamics thought experiment often called Maxwell's demons. For more information on its original implementation and origins click [here](https://pubmed.ncbi.nlm.nih.gov/9873902/). 

Below we will: 
1. Prepocess the images using histogram image matching. Various parameters are set to perform this.
2. Parameters, such as iterations and gaussian smoothing $\sigma$, are set for our demons registration
3. Create an instance of histogram matching image filter and execute with appropriate parameters
4. An instance of the demons registration filter is then created and parameters are assigned
5. The demons algorithm is then executed for our images. This will create a displacement field that will be updated as the similarity metric (cost function) is minimized. We will see the cost function values print for each iteration
6. Upon completion, our final displacement field will be used to create our overall transformation, outTx
7. Lastly, we can apply the transform to the moving image and resample.


```python
cost = []

print_flag = 1 # 1 to print cost func at each iteration, 0 to suppress.
DEMON_flag = 0 # 0 for classic demons and 1 (soft constraint) is for fast symmetric demons (hard constraint)
# 1. Initialize parameters for histogram matching
hist_levels = 1024
match_points = 7
matcher = sitk.HistogramMatchingImageFilter()

# 2. Initialize parameters for Demons
its = 500 # iterations for demons
sig = 1.0 # standard deviation 
# 3. Assign params and execute histogram match filter
matcher.SetNumberOfHistogramLevels(hist_levels)
matcher.SetNumberOfMatchPoints(match_points)
matcher.ThresholdAtMeanIntensityOn()
moving = matcher.Execute(moving, fixed)

# 4. Assign params and execute demons registration image filter
if DEMON_flag == 0:
    demons = sitk.DemonsRegistrationFilter()
else:
    demons = sitk.FastSymmetricForcesDemonsRegistrationFilter()
    
demons.SetNumberOfIterations(its)
demons.SetStandardDeviations(sig)
demons.AddCommand(sitk.sitkIterationEvent, lambda: command_iteration(demons,print_flag))
# 5. Run demons
displacementField = demons.Execute(fixed, moving)
print("-------")
print(f"Number Of Iterations: {demons.GetElapsedIterations()}")
print(f" RMS: {demons.GetRMSChange()}")

# 6. Create overall transform 
outTx = sitk.DisplacementFieldTransform(displacementField)
cost.append(demons.GetMetric())

# 7. Resample the image 
resampler = sitk.ResampleImageFilter()
resampler.SetReferenceImage(fixed)
resampler.SetInterpolator(sitk.sitkLinear)
resampler.SetDefaultPixelValue(100)
resampler.SetTransform(outTx)
out = resampler.Execute(moving)
#
print(demons)
```

### Display Results
Now that we have performed the registration, we can display the results!

```python
fig, axs = plt.subplots(2,3,figsize = (20,20))
fs = 24
its = np.linspace(1,demons.GetElapsedIterations(),num=demons.GetElapsedIterations())
img_fix = sitk.GetArrayFromImage(fixed)
img_mov = sitk.GetArrayFromImage(moving)
img_out = sitk.GetArrayFromImage(out)

axs[0,0].imshow(img_fix,cmap='gray'); axs[0,0].axis('off'); axs[0,0].set_title('Liver Expiration',fontsize=fs)
axs[0,1].imshow(img_mov,cmap='gray'); axs[0,1].axis('off'); axs[0,1].set_title('Liver Inspiration',fontsize=fs)
axs[0,2].imshow(img_out,cmap='gray'); axs[0,2].axis('off'); axs[0,2].set_title('Transformed Liver',fontsize=fs)
axs[1,0].imshow(img_fix-img_mov,cmap='gray'); axs[1,0].axis('off'); axs[1,0].set_title('Fixed-Moving',fontsize=fs)
axs[1,1].imshow(img_fix-img_out,cmap='gray'); axs[1,1].axis('off'); axs[1,1].set_title('Fixed-Transformed',fontsize=fs)

axs[1,2].plot(cost,'r',linewidth=3.0);
axs[1,2].set_title('Demons Cost Function Evolution',fontsize=fs);axs[1,2].set_xlabel('Iteration',fontsize=14);
axs[1,2].set_ylabel("Cost function",fontsize=14); 


plt.show()
print('Final Cost function value: ', cost[-1])
```
